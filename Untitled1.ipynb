```python
import os
import pandas as pd
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
from PIL import Image
import torchvision.transforms as transforms

# Load and sample the data
train_df = pd.read_csv('dataset/train.csv')
test_df = pd.read_csv('dataset/test.csv')

train_sample = train_df.sample(n=8000, random_state=42)
test_sample = test_df.sample(n=2000, random_state=42)

# Download images (assuming you've already done this)
download_folder = 'downloaded_images'
```

```python
class ProductImageDataset(Dataset):
    def __init__(self, dataframe, img_dir, transform=None):
        self.dataframe = dataframe
        self.img_dir = img_dir
        self.transform = transform
        self.entity_names = dataframe['entity_name'].unique()
        self.entity_to_idx = {entity: idx for idx, entity in enumerate(self.entity_names)}

    def __len__(self):
        return len(self.dataframe)

    def __getitem__(self, idx):
        img_name = os.path.join(self.img_dir, os.path.basename(self.dataframe.iloc[idx]['image_link']))
        image = Image.open(img_name).convert('RGB')

        if self.transform:
            image = self.transform(image)

        entity_name = self.dataframe.iloc[idx]['entity_name']
        entity_idx = self.entity_to_idx[entity_name]

        if 'entity_value' in self.dataframe.columns:
            entity_value = self.dataframe.iloc[idx]['entity_value']
            return image, entity_idx, entity_value
        else:
            return image, entity_idx

# Define transforms
transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
])

# Create datasets
train_dataset = ProductImageDataset(train_sample, download_folder, transform)
test_dataset = ProductImageDataset(test_sample, download_folder, transform)

# Create dataloaders
train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=32, shuffle=False)
```

```python
class CustomCNN(nn.Module):
    def __init__(self, num_classes):
        super(CustomCNN, self).__init__()
        self.features = nn.Sequential(
            nn.Conv2d(3, 64, kernel_size=3, padding=1),
            nn.ReLU(inplace=True),
            nn.MaxPool2d(kernel_size=2, stride=2),
            nn.Conv2d(64, 128, kernel_size=3, padding=1),
            nn.ReLU(inplace=True),
            nn.MaxPool2d(kernel_size=2, stride=2),
            nn.Conv2d(128, 256, kernel_size=3, padding=1),
            nn.ReLU(inplace=True),
            nn.MaxPool2d(kernel_size=2, stride=2),
        )
        self.classifier = nn.Sequential(
            nn.Linear(256 * 28 * 28, 512),
            nn.ReLU(inplace=True),
            nn.Dropout(),
            nn.Linear(512, num_classes)
        )
        self.regressor = nn.Sequential(
            nn.Linear(256 * 28 * 28, 512),
            nn.ReLU(inplace=True),
            nn.Dropout(),
            nn.Linear(512, 1)
        )

    def forward(self, x):
        x = self.features(x)
        x = x.view(x.size(0), -1)
        class_output = self.classifier(x)
        value_output = self.regressor(x)
        return class_output, value_output

# Initialize the model
num_classes = len(train_dataset.entity_names)
model = CustomCNN(num_classes)
```

```python
criterion_class = nn.CrossEntropyLoss()
criterion_value = nn.MSELoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader
from tqdm import tqdm
import re

def extract_numeric_value(value_str):
    # Extract the numeric part from the string
    match = re.search(r'(\d+(\.\d+)?)', value_str)
    if match:
        return float(match.group(1))
    return 0.0  # Default value if no numeric part is found

def train_model(model, train_loader, criterion_class, criterion_value, optimizer, num_epochs=10, device='cuda'):
    model.to(device)

    for epoch in range(num_epochs):
        model.train()
        running_loss = 0.0

        # Use tqdm for a progress bar
        pbar = tqdm(train_loader, desc=f'Epoch {epoch+1}/{num_epochs}')

        for images, entity_idxs, entity_values in pbar:
            images = images.to(device)
            entity_idxs = entity_idxs.to(device)

            # Process entity_values
            numeric_values = torch.tensor([extract_numeric_value(val) for val in entity_values]).float().to(device)

            optimizer.zero_grad()

            # Forward pass
            class_outputs, value_outputs = model(images)

            # Compute loss
            loss_class = criterion_class(class_outputs, entity_idxs)
            loss_value = criterion_value(value_outputs.squeeze(), numeric_values)
            loss = loss_class + loss_value

            # Backward pass and optimize
            loss.backward()
            optimizer.step()

            running_loss += loss.item()

            # Update progress bar
            pbar.set_postfix({'loss': loss.item()})

        epoch_loss = running_loss / len(train_loader)
        print(f'Epoch [{epoch+1}/{num_epochs}], Loss: {epoch_loss:.4f}')

# Set up loss functions and optimizer
criterion_class = nn.CrossEntropyLoss()
criterion_value = nn.MSELoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

# Train the model
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
train_model(model, train_loader, criterion_class, criterion_value, optimizer, num_epochs=10, device=device)
```

```python
ALLOWED_UNITS = {
    'item_weight': ['gram', 'kilogram', 'ounce', 'pound'],
    'item_volume': ['millilitre', 'litre', 'cubic centimetre', 'fluid ounce', 'gallon'],
    'voltage': ['volt'],
    'wattage': ['watt'],
    'maximum_weight_recommendation': ['kilogram', 'pound', 'ton'],
    'height': ['centimetre', 'inch', 'metre', 'foot'],
    'depth': ['centimetre', 'inch', 'metre', 'foot'],
    'width': ['centimetre', 'inch', 'metre', 'foot'],
}

# Define abbreviations for units if needed
UNIT_ABBREVIATIONS = {
    'gram': 'g',
    'kilogram': 'kg',
    'ounce': 'oz',
    'pound': 'lb',
    'millilitre': 'ml',
    'litre': 'L',
    'cubic_centimetre': 'cc',
    'fluid_ounce': 'fl oz',
    'gallon': 'gal',
    'volt': 'V',
    'watt': 'W',
    'centimetre': 'cm',
    'inch': 'in',
    'metre': 'm',
    'ton': 't',
    'foot': 'ft',
}
```

```python
def predict(model, test_loader, device):
    model.eval()
    predictions = []

    with torch.no_grad():
        for images, entity_idxs in test_loader:
            images = images.to(device)

            class_outputs, value_outputs = model(images)

            # Get the predicted class
            _, predicted_classes = torch.max(class_outputs, 1)

            # Get the predicted values
            predicted_values = value_outputs.squeeze()

            # Combine predictions
            for idx, val in zip(predicted_classes, predicted_values):
                entity_name = test_dataset.entity_names[idx]
                predictions.append((entity_name, val.item()))

    return predictions
```

```python
import math

def format_prediction(entity_name, value):
    if math.isnan(value) or value <= 0:
        return ""  # Return empty string for invalid predictions

    allowed_units = ALLOWED_UNITS.get(entity_name, [])
    if not allowed_units:
        return ""  # Return empty string if no allowed units for this entity type

    # Choose the appropriate unit based on the value
    unit = choose_unit(entity_name, value)

    # Convert the value to the chosen unit
    converted_value = convert_value(value, entity_name, unit)

    # Format the value to avoid scientific notation and round to 2 decimal places
    formatted_value = f"{converted_value:.2f}".rstrip('0').rstrip('.')

    # Use abbreviation if available, otherwise use full unit name
    unit_display = UNIT_ABBREVIATIONS.get(unit, unit.replace('_', ' '))

    return f"{formatted_value} {unit_display}"

def choose_unit(entity_name, value):
    # This function chooses the appropriate unit based on the value and entity type
    # You may need to adjust these thresholds based on your specific requirements
    if entity_name == 'item_weight':
        return 'gram' if value < 1000 else 'kilogram'
    elif entity_name == 'item_volume':
        return 'millilitre' if value < 1000 else 'litre'
    elif entity_name in ['height', 'depth', 'width']:
        return 'centimetre' if value < 100 else 'metre'
    else:
        return ALLOWED_UNITS[entity_name][0]  # Default to the first allowed unit

def convert_value(value, entity_name, target_unit):
    # This function converts the value to the target unit
    # You'll need to implement the conversion logic based on your specific requirements
    # Here's a simple example:
    if entity_name == 'item_weight':
        if target_unit == 'kilogram':
            return value / 1000
    elif entity_name == 'item_volume':
        if target_unit == 'litre':
            return value / 1000
    elif entity_name in ['height', 'depth', 'width']:
        if target_unit == 'metre':
            return value / 100
    return value  # Default case: no conversion needed
```

```python
import pandas as pd

# Make predictions
model = model.to(device)
predictions = predict(model, test_loader, device)

# Format predictions
formatted_predictions = [format_prediction(entity, value) for entity, value in predictions]

# Create submission dataframe
submission_df = pd.DataFrame({
    'index': test_sample['index'],
    'prediction': formatted_predictions
})

# Save submission file
submission_path = 'submission.csv'
submission_df.to_csv(submission_path, index=False)
print(f"Submission file saved to {submission_path}")
```
